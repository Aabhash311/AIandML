## 4.1 Exercise on Functions

### Task - 1: Unit Converter

```python
def convert_length(value, direction):
    """
    Converts length between meters and feet.

    Parameters:
        value (float): The numeric value to convert.
        direction (str): 'm_to_ft' or 'ft_to_m'.

    Returns:
        float: The converted length value.
    """
    if direction == 'm_to_ft':
        return value * 3.28084
    elif direction == 'ft_to_m':
        return value / 3.28084


def convert_weight(value, direction):
    """
    Converts weight between kilograms and pounds.

    Parameters:
        value (float): The numeric value to convert.
        direction (str): 'kg_to_lbs' or 'lbs_to_kg'.

    Returns:
        float: The converted weight value.
    """
    if direction == 'kg_to_lbs':
        return value * 2.20462
    elif direction == 'lbs_to_kg':
        return value / 2.20462


def convert_volume(value, direction):
    """
    Converts volume between liters and gallons.

    Parameters:
        value (float): The numeric value to convert.
        direction (str): 'l_to_gal' or 'gal_to_l'.

    Returns:
        float: The converted volume value.
    """
    if direction == 'l_to_gal':
        return value * 0.264172
    elif direction == 'gal_to_l':
        return value / 0.264172


def unit_converter():
    """
    Main function that prompts user to choose a conversion type
    (length, weight, or volume), input a value, and displays the result.
    Handles invalid input using try-except.

    Returns:
        float: The converted result, or None for invalid input.
    """
    print("Choose conversion type:")
    print("1. Length (meters <-> feet)")
    print("2. Weight (kg <-> lbs)")
    print("3. Volume (liters <-> gallons)")

    try:
        choice = input("Enter 1, 2, or 3: ").strip()

        if choice == '1':
            direction = input("Enter direction (m_to_ft / ft_to_m): ").strip()
            value = float(input("Enter value: "))
            result = convert_length(value, direction)
            print(f"Result: {result:.4f}")

        elif choice == '2':
            direction = input("Enter direction (kg_to_lbs / lbs_to_kg): ").strip()
            value = float(input("Enter value: "))
            result = convert_weight(value, direction)
            print(f"Result: {result:.4f}")

        elif choice == '3':
            direction = input("Enter direction (l_to_gal / gal_to_l): ").strip()
            value = float(input("Enter value: "))
            result = convert_volume(value, direction)
            print(f"Result: {result:.4f}")

        else:
            print("Invalid choice. Please enter 1, 2, or 3.")
            return None

        return result

    except ValueError:
        print("Invalid input. Please enter a numeric value.")
        return None


# Call the function
unit_converter()
```

### Task - 2: Mathematical Operations on a List

```python
def compute_sum(numbers):
    """
    Computes the sum of a list of numbers.

    Parameters:
        numbers (list): A list of numeric values.

    Returns:
        float: The sum of all elements.
    """
    total = 0
    for n in numbers:
        total += n
    return total


def compute_average(numbers):
    """
    Computes the average of a list of numbers.

    Parameters:
        numbers (list): A list of numeric values.

    Returns:
        float: The average of all elements.
    """
    if len(numbers) == 0:
        raise ValueError("Cannot compute average of an empty list.")
    return compute_sum(numbers) / len(numbers)


def compute_maximum(numbers):
    """
    Finds the maximum value in a list of numbers.

    Parameters:
        numbers (list): A list of numeric values.

    Returns:
        float: The maximum element.
    """
    if len(numbers) == 0:
        raise ValueError("Cannot find maximum of an empty list.")
    max_val = numbers[0]
    for n in numbers:
        if n > max_val:
            max_val = n
    return max_val


def compute_minimum(numbers):
    """
    Finds the minimum value in a list of numbers.

    Parameters:
        numbers (list): A list of numeric values.

    Returns:
        float: The minimum element.
    """
    if len(numbers) == 0:
        raise ValueError("Cannot find minimum of an empty list.")
    min_val = numbers[0]
    for n in numbers:
        if n < min_val:
            min_val = n
    return min_val


def math_operations():
    """
    Prompts user to choose a mathematical operation and input a list of numbers,
    then performs and displays the result. Handles invalid input with try-except.

    Returns:
        float: The result of the selected operation, or None on error.
    """
    print("Choose an operation:")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")

    try:
        choice = input("Enter 1, 2, 3, or 4: ").strip()
        raw = input("Enter numbers separated by spaces: ")
        numbers = [float(x) for x in raw.strip().split()]

        if len(numbers) == 0:
            raise ValueError("The list is empty.")

        if choice == '1':
            result = compute_sum(numbers)
            print(f"Sum: {result}")
        elif choice == '2':
            result = compute_average(numbers)
            print(f"Average: {result}")
        elif choice == '3':
            result = compute_maximum(numbers)
            print(f"Maximum: {result}")
        elif choice == '4':
            result = compute_minimum(numbers)
            print(f"Minimum: {result}")
        else:
            print("Invalid choice.")
            return None

        return result

    except ValueError as e:
        print(f"Error: {e}")
        return None


# Call the function
math_operations()
```

## 4.2 Exercise on List Manipulation

```python
# 1. Extract Every Other Element
def extract_every_other(lst):
    """Returns every other element starting from index 0."""
    return lst[::2]

print(extract_every_other([1, 2, 3, 4, 5, 6]))  # Expected: [1, 3, 5]
```

```python
# 2. Slice a Sublist
def get_sublist(lst, start, end):
    """Returns a sublist from start to end (inclusive)."""
    return lst[start:end + 1]

print(get_sublist([1, 2, 3, 4, 5, 6], 2, 4))  # Expected: [3, 4, 5]
```

```python
# 3. Reverse a List Using Slicing
def reverse_list(lst):
    """Returns a reversed version of the list using slicing."""
    return lst[::-1]

print(reverse_list([1, 2, 3, 4, 5]))  # Expected: [5, 4, 3, 2, 1]
```

```python
# 4. Remove the First and Last Elements
def remove_first_last(lst):
    """Returns the list without the first and last elements."""
    return lst[1:-1]

print(remove_first_last([1, 2, 3, 4, 5]))  # Expected: [2, 3, 4]
```

```python
# 5. Get the First n Elements
def get_first_n(lst, n):
    """Returns the first n elements of the list."""
    return lst[:n]

print(get_first_n([1, 2, 3, 4, 5], 3))  # Expected: [1, 2, 3]
```

```python
# 6. Extract Elements from the End
def get_last_n(lst, n):
    """Returns the last n elements of the list."""
    return lst[-n:]

print(get_last_n([1, 2, 3, 4, 5], 2))  # Expected: [4, 5]
```

```python
# 7. Extract Elements in Reverse Order (every second, from second-to-last)
def reverse_skip(lst):
    """Returns every second element starting from the second-to-last, moving backward."""
    return lst[-2::-2]

print(reverse_skip([1, 2, 3, 4, 5, 6]))  # Expected: [5, 3, 1]
```

## 4.3 Exercise on Nested List

```python
# 1. Flatten a Nested List
def flatten(lst):
    """Flattens a one-level nested list into a single list."""
    result = []
    for sublist in lst:
        for item in sublist:
            result.append(item)
    return result

print(flatten([[1, 2], [3, 4], [5]]))  # Expected: [1, 2, 3, 4, 5]
```

```python
# 2. Accessing Nested List Elements
def access_nested_element(lst, indices):
    """Returns the element at the given list of indices in a nested list."""
    element = lst
    for idx in indices:
        element = element[idx]
    return element

lst = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(access_nested_element(lst, [1, 2]))  # Expected: 6
```

```python
# 3. Sum of All Elements in a Nested List
def sum_nested(lst):
    """Recursively computes the sum of all elements in a nested list."""
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total

print(sum_nested([[1, 2], [3, [4, 5]], 6]))  # Expected: 21
```

```python
# 4. Remove Specific Element from a Nested List
def remove_element(lst, elem):
    """Removes all occurrences of elem from a nested list."""
    return [item for item in
            [[x for x in sublist if x != elem] for sublist in lst]]

print(remove_element([[1, 2], [3, 2], [4, 5]], 2))  # Expected: [[1], [3], [4, 5]]
```

```python
# 5. Find the Maximum Element in a Nested List
def find_max(lst):
    """Recursively finds the maximum element in a nested list."""
    max_val = None
    for item in lst:
        if isinstance(item, list):
            candidate = find_max(item)
        else:
            candidate = item
        if max_val is None or candidate > max_val:
            max_val = candidate
    return max_val

print(find_max([[1, 2], [3, [4, 5]], 6]))  # Expected: 6
```

```python
# 6. Count Occurrences of an Element in a Nested List
def count_occurrences(lst, elem):
    """Counts how many times elem appears in a nested list."""
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count

print(count_occurrences([[1, 2], [2, 3], [2, 4]], 2))  # Expected: 3
```

```python
# 7. Flatten a List of Lists of Lists (Deep Flatten)
def deep_flatten(lst):
    """Recursively flattens a deeply nested list into a single list."""
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result

print(deep_flatten([[[1, 2], [3, 4]], [[5, 6], [7, 8]]]))  # Expected: [1,2,3,4,5,6,7,8]
```

```python
# 8. Nested List Average
def average_nested(lst):
    """Computes the average of all elements in a nested list."""
    flat = deep_flatten(lst)
    return sum(flat) / len(flat)

print(average_nested([[1, 2], [3, 4], [5, 6]]))  # Expected: 3.5
```

### Problem - 4: Matrix Operations

```python
import numpy as np

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# 1. Prove A * A_inv = I
A_inv = np.linalg.inv(A)
product_A_Ainv = np.dot(A, A_inv)
print("1. A * A_inv (should be Identity):\n", np.round(product_A_Ainv, decimals=10))

# 2. Prove AB != BA
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\n2. AB:\n", AB)
print("BA:\n", BA)
print("AB == BA?", np.allclose(AB, BA))

# 3. Prove (AB)^T = B^T * A^T
AB_T = AB.T
BT_AT = np.dot(B.T, A.T)
print("\n3. (AB)^T:\n", AB_T)
print("B^T * A^T:\n", BT_AT)
print("(AB)^T == B^T * A^T?", np.allclose(AB_T, BT_AT))
```

```python
import numpy as np

# Solve system of equations:
# 2x - 3y + z  = -1
# x  - y  + 2z = -3
# 3x + y  - z  =  9

A_sys = np.array([[2, -3, 1],
                  [1, -1, 2],
                  [3,  1, -1]])

B_sys = np.array([-1, -3, 9])

# Method 1: Inverse Method (X = A_inv * B)
A_sys_inv = np.linalg.inv(A_sys)
X_inverse = np.dot(A_sys_inv, B_sys)
print("Solution using Inverse Method:")
print(f"x = {X_inverse[0]:.4f}, y = {X_inverse[1]:.4f}, z = {X_inverse[2]:.4f}")

# Method 2: np.linalg.solve
X_solve = np.linalg.solve(A_sys, B_sys)
print("\nSolution using np.linalg.solve:")
print(f"x = {X_solve[0]:.4f}, y = {X_solve[1]:.4f}, z = {X_solve[2]:.4f}")
```

## 10.2 Experiment: How Fast is NumPy?

```python
import numpy as np
import time

SIZE = 1_000_000

list_a = list(range(SIZE))
list_b = list(range(SIZE))
np_a = np.array(list_a)
np_b = np.array(list_b)

# 1. Element-wise Addition
start = time.time()
result_list = [list_a[i] + list_b[i] for i in range(SIZE)]
print(f"1. Addition - Python List:  {time.time() - start:.5f} seconds")

start = time.time()
result_np = np_a + np_b
print(f"   Addition - NumPy Array:  {time.time() - start:.5f} seconds")
```

```python
# 2. Element-wise Multiplication
start = time.time()
result_list = [list_a[i] * list_b[i] for i in range(SIZE)]
print(f"2. Multiplication - Python List:  {time.time() - start:.5f} seconds")

start = time.time()
result_np = np_a * np_b
print(f"   Multiplication - NumPy Array:  {time.time() - start:.5f} seconds")
```

```python
# 3. Dot Product
start = time.time()
dot_list = sum(list_a[i] * list_b[i] for i in range(SIZE))
print(f"3. Dot Product - Python List:  {time.time() - start:.5f} seconds")

start = time.time()
dot_np = np.dot(np_a, np_b)
print(f"   Dot Product - NumPy Array:  {time.time() - start:.5f} seconds")
```

```python
# 4. Matrix Multiplication (1000x1000)
MAT_SIZE = 1000
mat_list_a = [[float(i * MAT_SIZE + j) for j in range(MAT_SIZE)] for i in range(MAT_SIZE)]
mat_list_b = [[float(i * MAT_SIZE + j) for j in range(MAT_SIZE)] for i in range(MAT_SIZE)]

start = time.time()
result_mat = [[sum(mat_list_a[i][k] * mat_list_b[k][j]
               for k in range(MAT_SIZE))
               for j in range(MAT_SIZE)]
               for i in range(10)]  # Only 10 rows to avoid very long wait
list_time = time.time() - start
print(f"4. Matrix Multiply (10 rows) - Python List:  {list_time:.5f} seconds")

mat_np_a = np.random.rand(MAT_SIZE, MAT_SIZE)
mat_np_b = np.random.rand(MAT_SIZE, MAT_SIZE)

start = time.time()
result_np_mat = np.dot(mat_np_a, mat_np_b)
np_time = time.time() - start
print(f"   Matrix Multiply (1000x1000) - NumPy Array:  {np_time:.5f} seconds")
print(f"   NumPy is vastly faster for full matrix multiplication!")
```
